```python
import cv2
import pytesseract
```

```python
#Display image

from IPython.display import Image
Image(filename='1.png', width="400", height="200") 
```

```python
#Result of Image to Text

img = cv2.imread('1.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(pytesseract.image_to_string(img))
```
